```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv("koonjh.csv")

features = ['Cherry', 'Orange','Water Melon', 'Mago', 'Banana', 'Kiwi', 'Grape','Lemon', 'Melon', 'Peach', 'Tomato', 'Coconut', 'Potato']

def combine_features(row):
    return str(row['Cherry'])+" "+str(row['Orange'])+" "+str(row['Water Melon'])+" "+str(row['Mago'])+" "+str(row['Banana'])+" "+str(row['Kiwi'])+" "+str(row['Grape'])+" "+str(row['Lemon'])+" "+str(row['Melon'])+" "+str(row['Peach'])+" "+str(row['Tomato'])+" "+str(row['Coconut'])+" "+str(row['Potato'])

for i in features:
    data[i] = data[i].fillna(0) #filling all NaNs with blank string

data["combined_features"] = data.apply(combine_features,axis=1)
data.iloc[0].combined_features

cv = CountVectorizer()
count_matrix = cv.fit_transform(data["combined_features"])
cosine_sim = cosine_similarity(count_matrix)

def get_title_from_index(index):
    return data[data.index == index]["Cherry"].values[0]
def get_index_from_title(title):
    return data[data.title == title]["Mago"].values[0]

movie_user_likes = input("Enter name of the fruit you like most, | 'Please use a suitable name i.e. ['Cherry', 'OrangeWater Melon', 'Mago', 'Banana', 'Kiwi', 'Grape','Lemon', 'Melon', 'Peach', 'Tomato', 'Coconut', 'Potato']' : ")

try:
    movie_index = get_index_from_title(movie_user_likes)
    similar_movies = list(enumerate(cosine_sim[movie_index]))
    sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
    i=0
    print("Top similar to "+movie_user_likes+" are:\n")
    for element in sorted_similar_movies:
        print(get_title_from_index(element[0]))
        i+=1
        if i>5:
            break
except:
    print("Please enter a valid name.")
```
